You can concatenate tuples using the + operator to produce longer tuples:
In [18]: (4, None, 'foo') + (6, 0) + ('bar',)
Out[18]: (4, None, 'foo', 6, 0, 'bar')
Multiplying a tuple by an integer, as with lists, has the effect of concatenating that
many copies of the tuple:
In [19]: ('foo', 'bar') * 4
Out[19]: ('foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'bar')
Note that the objects themselves are not copied, only the references to them.

# Positional Arguments

## Definition
Arguments are passed to a function according to their position.

## Example

```python
def greet(name, age):
    print("Name:", name)
    print("Age:", age)

# Positional arguments
greet("Piyush", 21)
```

## Output

```python
Name: Piyush
Age: 21
```

## Key Points
- Matched by position.
- Order matters.
- Simple and concise.
- Incorrect order can produce unexpected results.

### Example

```python
greet(21, "Piyush")
```

Output:

```python
Name: 21
Age: Piyush
```

---

# Keyword Arguments

## Definition
Arguments are passed using parameter names.

## Example

```python
def greet(name, age):
    print("Name:", name)
    print("Age:", age)

# Keyword arguments
greet(age=21, name="Piyush")
```

## Output

```python
Name: Piyush
Age: 21
```

## Key Points
- Matched by parameter names.
- Order does not matter.
- Improves readability.
- Useful when functions have many parameters.

### Example

```python
greet(name="Piyush", age=21)
```

Output:

```python
Name: Piyush
Age: 21
```

---

# Mixing Positional and Keyword Arguments

## Valid

```python
greet("Piyush", age=21)
```

## Invalid

```python
greet(name="Piyush", 21)
```

Reason: Positional arguments must come before keyword arguments.

---

# Comparison

| Positional Arguments | Keyword Arguments |
|---------------------|------------------|
| Matched by position | Matched by name |
| Order matters | Order does not matter |
| Shorter syntax | More readable |
| Used for simple calls | Useful for functions with many parameters |

````python
def func(x,y,z=1.5):
  if z > 1:
    return z * ( x + y )
  else:
    return z / (x + y)
````

````python
func(5,6,z=0.7)
````

````python
func(3.14, 7, 3.5)
````

````python
states = ["     Alabama ", "Georgia!", "Georgia", "georgia", "FlOrIda","south carolina##", "West virginia?"]
import re
def clean_strings(strings):
  result = []
  for value in strings:
      value = value.strip()
      value = re.sub("[!#?]", "", value)
      value = value.title()
      result.append(value)
  print(result)
````

````python
clean_strings(states)
````
