```python
import re
txt = "hello planet"
x = re.findall("^hello",txt)
if x:
  print("Yes, the string starts with hello")
else:
  print("No match")
```

```python
import re
txt = "hello planet"
x = re.findall("planet$",txt)
if x:
  print("Yes, the string ends with planet")
else:
  print("No match")
```

```python
import re
txt = "hello planet"
x = re.findall("he. *o",txt)
print(x)
```

```python
import re
txt = "hello planet"
x = re.findall("he.+o",txt)
print(x)
```

```python
import re
txt = "hello planet"
x = re.findall("he.?o",txt)
print(x)
```

```python
# search for a sequence that starts with "he" followed by 0 or 1 character and an 0
import re
txt = "hello planet"
x = re.findall("he.?o",txt)
print(x)
```

```python
#search for a sequence that starts with "he" followed exactly 2 (any) characters and an "o"
import re
txt = "hello planet"
x = re.findall("he.{2}o",txt)
print(x)
```
